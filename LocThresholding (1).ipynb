```python
import numpy as np 
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('panda.jpg',0)        
x =img.shape[0]
y =img.shape[1]
print(img.shape)
print(img.shape[0])
z=8
for i in range(1,z):
    roi1=img[0:int(x/i),0:int(y/i)]
    ret1,th1=cv2.threshold(roi1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

    roi2=img[50:int(x/i),50:int(y/i)]
    ret2,th2=cv2.threshold(roi2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    roi3=img[80:int(x/i),120:int(y/i)]
    ret3,th3=cv2.threshold(roi3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    roi4=img[100:int(x/i),150:int(y/i)]
    ret4,th4=cv2.threshold(roi3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    roi5=img[120:int(x/i),150:int(y/i)]
    ret5,th5=cv2.threshold(roi3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
  
    toplam=ret1+ret2+ret3+ret4+ret5
    sonuc=toplam/5
retsonuc,mythreshold = cv2.threshold(img,sonuc,255,cv2.THRESH_BINARY)
ret1,globalthreshold = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
retotsu,otsu=cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

print(retsonuc)
print(retotsu)

titles=["ORGNAL","GLOBAL","OTSU","MYTHRESHOLD"]
images=[img,globalthreshold,otsu,mythreshold]
plt.figure(figsize=(30,30))
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()

    
    
```
